```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('data/lena.png')
```

```python
img_perspective = img.copy()
h, w = img.shape[:2]
```

```python
# 設定四對點，並取得 perspective 矩陣
point1 = np.array([[60, 40], [420, 40], [420, 510], [60, 510]], dtype=np.float32)
point2 = np.array([[0, 80], [w, 120], [w, 430], [0, 470]], dtype=np.float32)
M = cv2.getPerspectiveTransform(point1, point2)
```

```python
img_perspective = cv2.warpPerspective(img, M, (w, h))
```

```python
cv2.circle(img, (60, 40), 5, 3)
cv2.circle(img, (420, 40), 5, 3)
cv2.circle(img, (420, 510), 5, 3)
cv2.circle(img, (60, 510), 5, 3)
cv2.arrowedLine(img, (60,40), (0,80), 3)
cv2.arrowedLine(img, (420,40), (w,120), 3)
cv2.arrowedLine(img, (420,510), (w,430), 3)
cv2.arrowedLine(img, (60,510), (0,470), 3)
```

```python
img_show = np.hstack((img, img_perspective))
```

```python
cv2.imshow('image perspective transform', img_show)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(15,8))
plt.imshow(img_show[:,:,::-1])
```
