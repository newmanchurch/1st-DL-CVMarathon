```python
import cv2 
from matplotlib import pyplot as plt
import numpy as np
img_path = '../homework/data/lena.png'
img = cv2.imread(img_path, cv2.IMREAD_COLOR)
```

```python
# show in new window
cv2.imshow('bgr', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# show inline
plt.figure(figsize=(7,7))
plt.imshow(img[:,:,::-1])
```

```python
# another way
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
